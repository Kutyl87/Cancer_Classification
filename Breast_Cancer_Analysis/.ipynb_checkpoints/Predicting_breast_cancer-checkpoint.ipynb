```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
df = pd.read_csv("data_cleaned.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(16,14))
sns.heatmap(df.corr(),annot=True)
```

```python
# plot_color_gradients("Sequential", ["Red","Black"])
plt.figure(figsize=(12,6),dpi=500)
corr_diagnosis = df.corrwith(df["diagnosis"])
corr_diagnosis = corr_diagnosis.sort_values()
corr_diagnosis.plot(kind="bar",color=["teal","indigo", "rebeccapurple"])
plt.xlabel("Parameters")
plt.ylabel("Correlation")
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
X = df.drop(columns=["diagnosis"])
y = df["diagnosis"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
log = LogisticRegressionCV(max_iter=5000)
log.fit(X_train, y_train)
log.C_
```

```python
log.get_params()
```

```python
coeffs = np.sort(log.coef_)
plt.figure(figsize=(12,8))
sns.barplot(x=list(range(len(coeffs[0]))), y=coeffs[0])
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient")
plt.show()
```

```python
y_pred = log.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
conf_matr = confusion_matrix(y_test,y_pred)
```

```python
ConfusionMatrixDisplay(conf_matr).plot()
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import PrecisionRecallDisplay,RocCurveDisplay
import sklearn
```

```python
RocCurveDisplay.from_estimator(log,X_test,y_test)
```

```python
PrecisionRecallDisplay.from_estimator(log,X_test,y_test)
```
